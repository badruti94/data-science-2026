```python
import pandas as pd
df = pd.read_csv('/content/drive/MyDrive/data-science-2026/telco_churn.csv')
print(df.shape)
print(df["Churn"].value_counts(normalize=True))

from sklearn.model_selection import train_test_split
# TODO: encoding fitur kategorikal (mis. pd.get_dummies)
df = pd.get_dummies(df, drop_first=True)
# TODO: pisahkan X (fitur) dan y (target = Churn)
X = df.drop("Churn_Yes", axis=1)
y = df["Churn_Yes"]
X_tr, X_te, y_tr, y_te = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)

from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_estimators=300, class_weight="balanced", random_state=42)
rf.fit(X_tr, y_tr)

from sklearn.metrics import classification_report, roc_auc_score
# TODO: hitung prediksi dan probabilitas
y_pred = rf.predict(X_te)
y_prob = rf.predict_proba(X_te)[:,1]
# TODO: tampilkan classification_report dan ROC-AUC
print("Classification Report")
print(classification_report(y_te, y_pred))

print("ROC-AUC Score :", roc_auc_score(y_te, y_prob))

# TODO: hitung probabilitas churn (predict_proba)
probabilitas_churn = pd.DataFrame({"Probabilitas_Churn": y_prob})
print(probabilitas_churn.head())
# TODO: tulis kesimpulan singkat (3-4 kalimat)
```

Kesimpulan

Model Random Forest ini sudah berhasil dibuat dan dilatih dengan baik, lengkap dengan antisipasi data yang tidak seimbang (imbalanced). Dari hasil classification report dan skor ROC-AUC, kita bisa langsung melihat seberapa akurat model ini dalam menebak pelanggan yang berpotensi pergi. Angka-angka metrik tersebut menjadi acuan penting untuk mengukur performa model sebelum benar-benar dipakai
