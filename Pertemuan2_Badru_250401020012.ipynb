```python
import pandas as pd
import numpy as np

# ① Muat Dataset Titanic
URL = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
df = pd.read_csv(URL)

# ② Eksplorasi Awal
print(df.shape, df.columns.tolist())
print(df.isnull().sum())

# ③ Filtering – Penumpang Wanita Kelas 1
wk1 = df[(df['Sex']=='female') & (df['Pclass']==1)]
print(wk1['Survived'].mean(), wk1['Age'].mean())

# ④ Groupby – Tingkat Keselamatan per Kelas
hasil = df.groupby('Pclass').agg(
    total=('PassengerId', 'count'),
    selamat=('Survived', 'mean'),
).round(3)
print(hasil)

# Jawaban Pertanyaan
# Q1 : berdasarkan output print(df.shape, df.columns.tolist()), jumlah baris : 891, kolom 12
# Q2 : kolom yang memiliki nilai kosong : Age, Cabin, Embarked
# Q3 : dari print(wk1['Survived'].mean() didapat 0.968
# Q4 : 34,6 tahun
# Q5 : dari output table group by hasilnya adalah 0.630
# Q6 : hasilnya adalah selamat: 0.630, 0.473, 0.242
```
